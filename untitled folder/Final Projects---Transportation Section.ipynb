```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
file_path = 'september-2023.xlsx'  # Change this to the path of your Excel file
data = pd.read_excel(file_path)

# Calculate average duration for each start and end station
avg_duration_start = data.groupby('Start Station Name')['Duration'].mean().sort_values(ascending=False)
avg_duration_end = data.groupby('End Station Name')['Duration'].mean().sort_values(ascending=False)

# Select the top 10 stations for visualization
top_start_stations = avg_duration_start.head(10)
top_end_stations = avg_duration_end.head(10)

# Plotting
fig, ax = plt.subplots(2, 1, figsize=(14, 12))
ax[0].bar(top_start_stations.index, top_start_stations.values, color='blue')
ax[0].set_title('Average Rental Duration by Start Station (Top 10)')
ax[0].set_ylabel('Average Duration (seconds)')
ax[0].set_xticklabels(top_start_stations.index, rotation=45, ha='right')

ax[1].bar(top_end_stations.index, top_end_stations.values, color='green')
ax[1].set_title('Average Rental Duration by End Station (Top 10)')
ax[1].set_ylabel('Average Duration (seconds)')
ax[1].set_xticklabels(top_end_stations.index, rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths for all datasets
file_paths = [
    'september-2023.xlsx',
    'october-2023.xlsx',
    'november-2023.xlsx',
    'december-2023.xlsx',
    'january-2024.xlsx',
    'february-2024.xlsx',
    'march-2024.xlsx'
]

# Load all data and calculate average duration for each station
all_data = pd.DataFrame()

for file in file_paths:
    # Load each file
    data = pd.read_excel(file)
    
    # Calculate average duration for each start and end station
    avg_start = data.groupby('Start Station Name')['Duration'].mean().reset_index()
    avg_start.rename(columns={'Duration': 'Avg Start Duration'}, inplace=True)
    
    avg_end = data.groupby('End Station Name')['Duration'].mean().reset_index()
    avg_end.rename(columns={'Duration': 'Avg End Duration'}, inplace=True)
    
    # Merge to all_data dataframe
    all_data = pd.concat([all_data, avg_start, avg_end])

# Combine all results and re-calculate average across months
all_data_start = all_data.groupby('Start Station Name')['Avg Start Duration'].mean().sort_values(ascending=False)
all_data_end = all_data.groupby('End Station Name')['Avg End Duration'].mean().sort_values(ascending=False)

# Select the top 10 stations for start and end from the combined data
top_10_start_stations = all_data_start.head(10)
top_10_end_stations = all_data_end.head(10)

# Plotting the top 10 stations for both start and end
fig, ax = plt.subplots(2, 1, figsize=(14, 12))

# Top 10 Start Stations
ax[0].bar(top_10_start_stations.index, top_10_start_stations.values, color='blue')
ax[0].set_title('Top 10 Stations by Average Start Duration')
ax[0].set_ylabel('Average Start Duration (seconds)')
ax[0].set_xticks(range(len(top_10_start_stations)))
ax[0].set_xticklabels(top_10_start_stations.index, rotation=45, ha='right')

# Top 10 End Stations
ax[1].bar(top_10_end_stations.index, top_10_end_stations.values, color='green')
ax[1].set_title('Top 10 Stations by Average End Duration')
ax[1].set_ylabel('Average End Duration (seconds)')
ax[1].set_xticks(range(len(top_10_end_stations)))
ax[1].set_xticklabels(top_10_end_stations.index, rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
