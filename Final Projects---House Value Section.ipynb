```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("value.csv")

df["Average"] = df.iloc[:, 1:].mean(axis=1)

df_sorted = df.sort_values(by="Average", ascending=False)

top_25 = df_sorted.head(25)
neighborhoods = top_25["Neighborhood"].to_numpy()
averages = top_25["Average"].to_numpy()

plt.figure(figsize=(12, 6))
plt.bar(neighborhoods, averages)
plt.xlabel("Neighborhood")
plt.ylabel("Average Value")
plt.title("Top 25 Neighborhoods (Average Value - Highest to Lowest)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

plt.show()
```
